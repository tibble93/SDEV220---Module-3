Given an array arr[] containing only 0s, 1s, and 2s. Sort the array in ascending order.

```python
def number_sort(arr):
    # Set counters for 0s, 1s, and 2s
    count_zeros: int = 0
    count_ones: int = 0
    count_twos: int = 0

    for number in arr:
        if number == 0:
            count_zeros += 1
        elif number == 1:
            count_ones += 1
        elif number == 2:
            count_twos += 1

    # The positional index will insert the sorted numbers
    positional_index: int = 0

    # Insert 0s, 1s, and 2s back into the array in sorted order
    for x in range(count_zeros):
        arr[positional_index] = 0
        positional_index += 1

    for x in range(count_ones):
        arr[positional_index] = 1
        positional_index += 1

    for x in range(count_twos):
        arr[positional_index] = 2
        positional_index += 1

#-----------------------------------------------------------------------------------
# Create a list of numbers 
number_list: list = [0, 1, 2, 0, 1, 2, 1, 0]
# Call the function "number_sort with the list as an argument
number_sort(number_list)
# Print the sorted list in ascending order
print(number_list)
```
